### Logistic Regression


```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, recall_score, precision_score
from sklearn.metrics import confusion_matrix
```

```python
#Multivariate Logistic Regression

model_multivariate = sm.Logit(y_train, X_train).fit()
print(model_multivariate.summary())

```

```python
# prompt: now take exponent of the coefficent

import numpy as np
# Get the coefficients
coefficients = model_multivariate.params

# Calculate the exponents of the coefficients
odds_ratios = np.exp(coefficients)

# Print the odds ratios
print(odds_ratios)
```

```python
# prompt: Predicting probabilities on train and test data

# Predict probabilities on train data
y_train_pred_prob = model_multivariate.predict(X_train)

# Predict probabilities on test data
y_test_pred_prob = model_multivariate.predict(X_test)

print("Predicted probabilities on train data:\n", y_train_pred_prob)
print("\nPredicted probabilities on test data:\n", y_test_pred_prob)
```

```python
# prompt: Predicting probabilities on train and test data compare actual and predicted

import pandas as pd
# Predict probabilities on train data
y_train_pred_prob = model_multivariate.predict(X_train)

# Predict probabilities on test data
y_test_pred_prob = model_multivariate.predict(X_test)

# Convert probabilities to binary predictions (0 or 1) using a threshold (e.g., 0.5)
threshold = 0.5
y_train_pred = (y_train_pred_prob >= threshold).astype(int)
y_test_pred = (y_test_pred_prob >= threshold).astype(int)

# Compare actual and predicted values for train data
train_comparison = pd.DataFrame({'Actual': y_train, 'Predicted': y_train_pred})
print("Train Data Comparison:")
print(train_comparison.head())

# Compare actual and predicted values for test data
test_comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})
print("\nTest Data Comparison:")
print(test_comparison.head())
```

```python
# prompt: create and plot confusion metrices

import matplotlib.pyplot as plt
# Confusion Matrix for Train Data
cm_train = confusion_matrix(y_train, y_train_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Train Data')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Confusion Matrix for Test Data
cm_test = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Test Data')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
# 0 No and 1 = Yes
```

```python
# prompt: calculate training and testing accuarcy, precission, recall and  f1 score

# Calculate metrics for train data
accuracy_train_LR = accuracy_score(y_train, y_train_pred)
precision_train_LR = precision_score(y_train, y_train_pred)
recall_train_LR = recall_score(y_train, y_train_pred)
f1_train_LR = f1_score(y_train, y_train_pred)

# Calculate metrics for test data
accuracy_test_LR = accuracy_score(y_test, y_test_pred)
precision_test_LR = precision_score(y_test, y_test_pred)
recall_test_LR = recall_score(y_test, y_test_pred)
f1_test_LR = f1_score(y_test, y_test_pred)

# Print the metrics
print("Train Data Metrics:")
print("Accuracy:", accuracy_train_LR)
print("Precision:", precision_train_LR)
print("Recall:", recall_train_LR)
print("F1-score:", f1_train_LR)

print("\nTest Data Metrics:")
print("Accuracy:", accuracy_test_LR)
print("Precision:", precision_test_LR)
print("Recall:", recall_test_LR)
print("F1-score:", f1_test_LR)
```

```python
# prompt: create a ROC plot for logistic regression

import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

# Calculate ROC curve and AUC for train data
fpr_train_LR, tpr_train_LR, thresholds_train = roc_curve(y_train, y_train_pred_prob)
roc_auc_train_LR = roc_auc_score(y_train, y_train_pred_prob)

# Calculate ROC curve and AUC for test data
fpr_test_LR, tpr_test_LR, thresholds_test = roc_curve(y_test, y_test_pred_prob)
roc_auc_test_LR = roc_auc_score(y_test, y_test_pred_prob)

# Plot ROC curves
plt.figure(figsize=(8, 6))
plt.plot(fpr_train_LR, tpr_train_LR, label='Train ROC curve (area = %0.2f)' % roc_auc_train_LR)
plt.plot(fpr_test_LR, tpr_test_LR, label='Test ROC curve (area = %0.2f)' % roc_auc_test_LR)
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line for random classifier
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```

### KNN(K nearest Neighbour)

```python
# prompt: now use KNN model in which i have define range of k values from (1 to 30). perform grid search for best k value then predict values on train and test. then calculate performance metrices

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Define the range of k values
k_range = range(1, 31)

# Create a KNN classifier object
knn = KNeighborsClassifier()

# Define the parameter grid for grid search
param_grid = dict(n_neighbors=k_range)

# Perform grid search to find the best k value
grid = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)

# Print the best k value
print("Best k:", grid.best_params_['n_neighbors'])

# Get the best KNN model
best_knn = grid.best_estimator_

# Predict on train and test sets
y_train_pred = best_knn.predict(X_train)
y_test_pred = best_knn.predict(X_test)

# Calculate performance metrics for training set
train_accuracy_knn = accuracy_score(y_train, y_train_pred)
train_precision_knn = precision_score(y_train, y_train_pred)
train_recall_knn = recall_score(y_train, y_train_pred)
train_f1_knn = f1_score(y_train, y_train_pred)

# Calculate performance metrics for test set
test_accuracy_knn = accuracy_score(y_test, y_test_pred)
test_precision_knn = precision_score(y_test, y_test_pred)
test_recall_knn = recall_score(y_test, y_test_pred)
test_f1_knn = f1_score(y_test, y_test_pred)

# Print performance metrics
print(f"Training Accuracy: {train_accuracy_knn:.2f}")
print(f"Training Precision: {train_precision_knn:.2f}")
print(f"Training Recall: {train_recall_knn:.2f}")
print(f"Training F1-Score: {train_f1_knn:.2f}")
print()
print(f"Test Accuracy: {test_accuracy_knn:.2f}")
print(f"Test Precision: {test_precision_knn:.2f}")
print(f"Test Recall: {test_recall_knn:.2f}")
print(f"Test F1-Score: {test_f1_knn:.2f}")
```
